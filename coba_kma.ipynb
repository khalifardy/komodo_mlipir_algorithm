```python
import numpy as np
from fungsi_benchmark import BenchmarkFunctions
from optimizer.KomodoMlipirAlgorithm import KMA
```

```python
benchmark = BenchmarkFunctions()
```

```python
search_space_optimizer = [(-5.12, 5.12), (-5.12, 5.12)] 
```

```python
optimizer = KMA(fitness_function = fitness_function,search_space = search_space_optimizer, max_iterations = 1000)
```

```python
optimizer.fit()
```

```python
optimizer.get_results()
```
